## Calculating Profitability
Assumes 180 days since TDate or current date. Should create separate database with only purchases and calc/graph profitability based on that.
Should be modified to ensure that only purchase dates are used.
180 days value should also be changed to reflect average holding period.

```python
TradeTypecounts = iaaii.TradeType.value_counts()
TradeTypecounts
```

```python
def calculate_profitability(row):
    stock_symbol = row['Ticker']
    trade_date = row['TDate']
    trade_type = row['TradeType']
    shares = row['Qty']

    try:
        if trade_type.startswith('P'):
            stock_data = yf.download(stock_symbol, start=trade_date, end=trade_date + timedelta(days=180), progress=False)
        elif trade_type.startswith('S'):
            stock_data = yf.download(stock_symbol, start=trade_date - timedelta(days=180), end=trade_date, progress=False)
        else:
            return pd.Series({'Percentage Profit': 0.0, 'Dollar Profit': 0.0})

        if stock_data.empty:
            return pd.Series({'Percentage Profit': 0.0, 'Dollar Profit': 0.0})

        closing_prices = stock_data['Close']

        if trade_type.startswith('P'):
            end_date = min(trade_date + timedelta(days=180), pd.to_datetime('2023-11-30'))
        else:
            end_date = max(trade_date - timedelta(days=180), pd.to_datetime('2023-01-01'))

        closing_prices = closing_prices[(closing_prices.index >= end_date) | (closing_prices.index <= end_date)]

        purchase_price = closing_prices.iloc[0]
        current_price = closing_prices.iloc[-1]
        percentage_profit = ((current_price - purchase_price) / purchase_price)
        dollar_profit = (current_price - purchase_price) * shares

    except Exception as e:
        #error_tickers.append(stock_symbol)
        print(f"problem downloading {stock_symbol}: {str(e)}")
        return pd.Series({'Percentage Profit': 0.0, 'Dollar Profit': 0.0})

    return pd.Series({'Percentage Profit': percentage_profit, 'Dollar Profit': dollar_profit})
```

```python
#profit_df = iaaii.apply(calculate_profitability, axis=1)
#iaaii = pd.concat([iaaii, profit_df], axis=1)
iaaii[['Percentage Profit', 'Dollar Profit']] = iaaii.apply(calculate_profitability, axis=1)
```
